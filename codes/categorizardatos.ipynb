```python
def tout_class(data): 
  TOUTC = [(-20, -9, 'A_Frígido'),
          (-9, 0, 'B_Helada'),
          (0, 7, 'C_MuyFrío'),
          (7, 13, 'D_Frío'),
          (13, 18, 'E_Fresco'),
          (18, 24, 'F_Cómodo'),
          (24, 29, 'G_Caliente'),
          (29, 35, 'H_MuyCaliente'),
          (35, 50, 'I_Tórrido')]
  data["tout_c"] = np.nan
  for i, j, k in TOUTC:
    rowIndex = data.loc[(data['tout'] >= i) & (data['tout'] < j)].index
    data.loc[rowIndex, 'tout_c'] = k

def rh_class(data):
  RHC = [(0, 20, 'A_MuySeco'),
        (20, 40, 'B_Seco'),
        (40, 60, 'C_Confort'),
        (60, 80, 'D_Húmedo'),
        (70, 100, 'C_MuyHúmedo')]
  data['rh_c'] = np.nan
  for i, j, k in RHC:
    rowIndex = data.loc[(data['rh'] >= i) & (data['rh'] < j)].index
    data.loc[rowIndex, 'rh_c'] = k

def wd_class(data):
  WDC = [(0, 22.5, 'N'), 
        (22.5, 67.5, 'NE'), 
        (67.5, 112.5, 'E'), 
        (112.5, 157.5, 'SE'), 
        (157.5, 202.5, 'S'), 
        (202.5, 247.5, 'SW'), 
        (247.5, 292.5, 'W'),
        (292.5, 337.5, 'NW'),
        (337.5, 361, 'N')]
  data['wd_c'] = np.nan
  for i, j, k in WDC:
    rowIndex = data.loc[(data['wd'] >= i) & (data['wd'] < j)].index
    data.loc[rowIndex, 'wd_c'] = k

def ws_class(data):
  WSC = [(0, 1, 'A_Calma'), 
        (1, 5, 'B_Ventolina'), 
        (5, 11, 'C_Brisa Muy Débil'), 
        (11, 19, 'D_Brisa Débil'), 
        (19, 29, 'E_Brisa Moderada'), 
        (29, 39, 'F_Brisa Fuerte'), 
        (39, 49, 'G_Brisa Muy Fuerte'),
        (49, 61, 'H_Vendaval Débil'),
        (61, 74, 'I_Vendaval'),
        (74, 88, 'J_Vendaval FUerte'), 
        (88, 102, 'K_Tormenta'), 
        (102, 117, 'L_Tormenta Violenta'), 
        (117, 134, 'M_Huracán')]
  data['ws_c'] = np.nan
  for i, j, k in WSC:
    rowIndex = data.loc[(data['ws'] >= i) & (data['ws'] < j)].index
    data.loc[rowIndex, 'ws_c'] = k

def sr_class(data):
  SRC = [(0, 0.2, 'A_Muy Baja'), 
        (0.2, 0.4, 'B_Baja'), 
        (0.4, 0.6, 'C_Media'), 
        (0.6, 0.8, 'D_Alta'), 
        (0.8, 1.1, 'E_Muy Alta')]
  data['sr_c'] = np.nan
  for i, j, k in SRC:
    rowIndex = data.loc[(data['sr'] >= i) & (data['sr'] < j)].index
    data.loc[rowIndex, 'sr_c'] = k

def prs_class(data):
  PRS = [(650, 700, 'A_Baja'), 
        (700, 720, 'B_Normal'), 
        (720, 730, 'C_Alta')]
  data['prs_c'] = np.nan
  for i, j, k in PRS:
    rowIndex = data.loc[(data['prs'] >= i) & (data['prs'] < j)].index
    data.loc[rowIndex, 'prs_c'] = k

```

```python
def save_to_Drive(path,df,name): 
  route = path+name+'.csv'
  with open(route, 'w', encoding = 'utf-8-sig') as f:
    df.to_csv(f, index=False)


def df_stats(data_stats): 
  data_stats = data_stats.select_dtypes(include=np.number)
 
  stats = pd.concat([
            data_stats.describe().T,
            data_stats.mode(numeric_only=True).iloc[0].rename("Mode"),
            data_stats.var( numeric_only=True).rename("Var"),
            data_stats.sem(numeric_only=True).rename('SE Mean'),
            data_stats.isnull().sum().rename('N*')
            ],
            axis = 1)
  stats = stats.rename(columns={"count": "N",
                            "std": "StdDev",
                            "25%":"Q1",
                            "75%":"Q3",
                            "50%":"Median",
                            "min":"Minimum",
                            "max":"Maximum",
                            "mean":"Mean"
                            })
  column_names = ["N",
                  "N*",
                  "Mean", 
                  "SE Mean", 
                  "StdDev", 
                  "Var", 
                  "Minimum", 
                  "Maximum", 
                  "Q1", 
                  "Median", 
                  "Q3", 
                  "Mode"]
  return stats.reindex(columns=column_names)
```
